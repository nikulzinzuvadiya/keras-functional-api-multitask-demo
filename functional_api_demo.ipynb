```python
# Importing required Keras functionality for creating models

from keras.models import Model # Functional API for model creation
```

```python
model = Model(inputs=x, outputs=[Output1, Output2])  # Define the model with one input and two outputs
```

```python
from keras.layers import * # Importing necessary layers like Dense, Input
```

```python
x = Input(shape=(3,))  # Defining the input layer with 3 features (shape=(3,))
# Explanation:
# - `Input` acts as a placeholder for the input data.
# - Input shape (3,) means the model expects 3 features as input.
```

```python
# Add Hidden Layers
hidden1 = Dense(128, activation='relu')(x)  # First hidden layer with 128 neurons and ReLU activation
hidden2 = Dense(64, activation='relu')(hidden1)  # Second hidden layer with 64 neurons and ReLU activation


#Define Multiple Outputs
Output1 = Dense(1, activation='linear')(hidden2)  # First output for regression (continuous value)
Output2 = Dense(1, activation='sigmoid')(hidden2)  # Second output for binary classification
```

```python
model.summary()
```

```python
from keras.utils import plot_model # For visualizing the model architecture

plot_model(model,show_shapes=True,show_layer_names=True)

# Explanation:
# - `plot_model` creates a graphical visualization of the model.
# - `show_shapes=True` displays the shape of inputs and outputs for each layer.
# - `show_layer_names=True` includes the names of layers in the diagram.
```
